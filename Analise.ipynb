```python
#Importando a biblioteca 
import pandas as pd
```

```python
#Ler conjunto de dados
df = pd.read_csv("gapminder.csv")
```

```python
#visualizando as 5 primeiras linhas
df.head()
```

```python
#Renomeando nome da coluna
df.rename(columns={"country":"Pais","year":"Ano" , "population":"Pop Total","continent":"Continente","life_exp":"Expectativa de Vida","gdp_cap":"PIB"})
```

```python
#head aceita parametros
df.head()
```

```python
#Shape para saber total de linhas e colunas
df.shape
```

```python
#columns Retorna nomes das colunas
df.columns
```

```python
#dtypes para saber tipo de dados de cada coluna
df.dtypes
```

```python
#tail para saber as ultimas linhas aceita parametros 
df.tail()
```

```python
#describe retorna informações estatísticas
df.describe()
```

```python
#unique retorna valores selecionado
df["Continente"].unique()
```

```python
#loc retorna valores expecifico
Oceania = df.loc[df["Continente"] == "Oceania"]
Oceania.head()
```

```python
# groupby Agrupamentos de dados
df.groupby("Continente")["Pais"].nunique()
```

```python
#Saber para cada ano a expectativa de vida média
df.groupby("Ano")["Expectativa de Vida"].mean()
```
